## Learning Julia 3D plotting capability

```julia
function coil(theta)
  return [cos(theta),sin(theta),0*theta]
end
```

```julia
l=linspace(0,2*pi,1000)
coils=reshape(coil(l,0),1000,3)
coils2=reshape(coil(l,1),1000,3)
```

```julia
plot3D(coils[1:1000,1],coils[1:1000,2],coils[1:1000,3])
plot3D(coils2[1:1000,1],coils2[1:1000,2],coils2[1:1000,3])
```

```julia
#Function takes in 3 vectors one for each x, y,z and crosses them by a single vector R in 3 dimensions
function threeCross(x,y,z,R)
  cx=x
  cy=x
  cz=x
  print(cx)
  for n=1:length(x)
    cr=cross([x[n],y[n],z[n]],R)
    cx[n]=cr[1]
    cy[n]=cr[2]
    cz[n]=cr[3]
  end
  return cx,cy,cz
end
```

```julia
#This function returns the displacement vector dl as three vectors
function dlVec(f,n,start, stop)
  dl=linspace(start,stop,n)
  dvx=zeros(Float64,length(dl),1)
  dvy=zeros(Float64,length(dl),1)
  dvz=zeros(Float64,length(dl),1)
  dv=(f(dl[2])-f(dl[1]))/2
  dvx[1]=dv[1]
  dvy[2]=dv[2]
  dvz[3]=dv[3]
  dv=(f(dl[length(dl)])-f(dl[length(dl)-1]))/2
  dvx[length(dl)]=dv[1]
  dvy[length(dl)]=dv[2]
  dvz[length(dl)]=dv[3]
  for n=2:(length(dl)-1)
    dv=(f(dl[n+1])-f(dl[n-1]))/2
    dvx[n]=dv[1]
    dvy[n]=dv[2]
    dvz[n]=dv[3]
  end
  return dvx,dvy,dvz
end
```

```julia
dx,dy,dz=dlVec(coil,360,1,2*pi)
```
