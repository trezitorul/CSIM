```julia
d20=0.0001294*2*pi
d24=0.00008125496*2*pi
```

```julia
using ImmutableArrays
c20=linSpiral(d20/(2*pi),0,.0254,250)
asb120=Assembly([translate(c20,Vector3(0.0,0.0,d20))])
asb220=Assembly([c20])
```

```julia
Energies=[]
for i=.1:.05:.5
    E=coilEnergy(asb120,asb220,i,50,.1,.008,[0,0,0],"LoadingBatch(.1to.5)3-27-14")
    println(E)
    Energies=vcat(Energies,E)
    println(Energies)
end
writecsv("../LoadingBatch(.1to.5)3-27-14/Energies.csv",Energies)
```

```julia
EV=[]
println("Executing Voltage Batch")
fileName="VoltageBatch(10to150)3-27-14"
for V=30:10:150
    E=coilEnergy(asb120,asb220,.1,V,.1,.008,[0,0,0],fileName)
    println(E)
    EV=vcat(EV,E)
    println(EV)
end
writecsv(string("../",fileName,".csv"),Energies)
```

```julia
EC=[]
println("Executing CapSize Batch")
fileName="CapSizeBatch(10to150)3-27-14"
for C=.001:.001:.01
    E=coilEnergy(asb120,asb220,.1,50,.1,C,[0,0,0],fileName)
    println(E)
    EC=vcat(EC,E)
    println(EC)
end
writecsv(string("../",fileName,".csv"),Energies)
```
