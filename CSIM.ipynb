```julia
#Function takes in 3 vectors one for each x, y,z and crosses them by a single vector R in 3 dimensions
function threeCross(x,y,z,R)
  cx=x
  cy=x
  cz=x
  print(cx)
  for n=1:length(x)
    cr=cross([x[n],y[n],z[n]],R)
    cx[n]=cr[1]
    cy[n]=cr[2]
    cz[n]=cr[3]
  end
  return cx,cy,cz
end
```

```julia
#This function returns the displacement vector dl as three vectors
function dlVec(f,n,start, stop)
  dl=linspace(start,stop,n)
  dvx=zeros(Float64,length(dl),1)
  dvy=zeros(Float64,length(dl),1)
  dvz=zeros(Float64,length(dl),1)
  dv=(f(dl[2])-f(dl[1]))/2
  dvx[1]=dv[1]
  dvy[2]=dv[2]
  dvz[3]=dv[3]
  dv=(f(dl[length(dl)])-f(dl[length(dl)-1]))/2
  dvx[length(dl)]=dv[1]
  dvy[length(dl)]=dv[2]
  dvz[length(dl)]=dv[3]
  for n=2:(length(dl)-1)
    dv=(f(dl[n+1])-f(dl[n-1]))/2
    dvx[n]=dv[1]
    dvy[n]=dv[2]
    dvz[n]=dv[3]
  end
  return dvx,dvy,dvz
end
```

```julia
#Generates a infintesimal difference matrix required to get dl in one dimension. 
function D(N)
  DM=-1*diagm(ones(N-1),-1)+diagm(ones(N-1),1)
  DM[1,1]=-1.0
  DM[N,N]=1.0
  return DM
end
```

```julia
#Generates multidimensional difference matrix required to get dl vect in dim dimensions 
function A(N,dim)
  return sparse(kron(eye(dim),D(N)))
end
```

```julia
#Returns the vector point from the function to the point R.
function rVec(func,dim,R)
  N=convert(Int64,length(func)/dim)
  RVEC=[R[1]*ones(N,1), R[2]*ones(N,1), R[3]*ones(N,1)]
  return RVEC-func
end
```

```julia
#Takes the crossProd of between the difference vector given a point r and the points on a line specified by dl
function crossProd(dl,rVec,dim)
  N=convert(Int64,length(dl)/dim)
  ndl=reshape(dl,N,dim)
  nrVec=reshape(rVec,N,dim)
  crP=zeros(Float64,N,dim)
  for n=1:N
    crP[n,1:dim]=reshape(cross(reshape(ndl[n,1:dim],dim),reshape(nrVec[n,1:dim],dim)),1,dim)
  end
  print(crP)
  return reshape(crP,N*dim)
end
  
```

```julia
#Simple 2D spiral with start radius of R0, and incremental size increase of dw for complete period of theta
function spiral(dw,R0,z,theta)
  r=dw*theta+R0
  return [r*cos(theta),r*sin(theta),z]
end
```

```julia
#Simple 2D spiral with start radius of R0, and incremental size increase of dw for complete period of theta
function dspiral(dw,R0,z,theta)
  r=dw*theta+R0
  return [dw*cos(theta)-r*sin(theta),dw*sin(theta)+r*cos(theta),0]
end
```

```julia
#This function generates the spirals in question, they are N independent spirals that have the following parameter list
#n=number of turns, R0 where they begin spiraling radially (always set at theta = 0) 
#RF=where the spiral stops spiralling, thetaS=where theta starts, thetaF where theta ends,
#This is represented for each leveln in params which is an array of arrays, each array consists of 
#[n,dw,R0], this is the parameter list. The limits on theta will be set by the optimizer not here.
function spiralMulti(params,theta)
  N=theta/pi
  N=floor(N)
  count=0
  for k=1:length(params)
    count=count+params[k][1]
    if(count>N)
      N=k
      break
    end
  end
  dz=1 #The height that each level is at (in cm)
  R0=params[N][3]
  dw=params[N][2]
  r=dw*(theta-N*theta)+R0
  return r*cos(theta),r*sin(theta),N*dz
end
```

```julia
#Converts vector of vector into len(A)x3 matrix
function converter(A)
  newA=[[testSpiral[n][1] for n=[1:length(testSpiral)]],[testSpiral[n][2] for n=[1:length(testSpiral)]],[testSpiral[n][3] for n=[1:length(testSpiral)]]]
  return reshape(newA,length(A),3)
end
```

```julia
mu=1
function biotSavart(I,f,df,r,theta)
  dl=df(theta)*theta
  dr=f(theta)-r
  ldr=sqrt(dr[1]^2+dr[2]^2+dr[3]^2)
  f=I*cross(dl,dr)/ldr^3
  return f
end
```

```julia
biotSavart(1,theta -> spiral(.01,0,1,theta),theta->dspiral(.01,0,1,theta),[0,0,0],50)
```

```julia
function B(I,f,df,r, thetaMax)
  IX=quadgk(theta -> biotSavart(I,f,df,r,theta),0,thetaMax)
  return IX
end
```

```julia
B(1,theta -> spiral(.01,0,1,theta),theta->dspiral(.01,0,1,theta),[0,0,0],240)
```
