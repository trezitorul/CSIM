## Misc Plots

```julia
using PyPlot
```

```julia
function capDischarge(I,t, tau)
    return I*exp(-t/tau)
end
```

```julia
t=[0:.01:10];
```

```julia
plot(t, capDischarge(1,t,1),color="black",linewidth=2)
xlabel("Time (sec)")
ylabel("I (Amps)")
title("Capacitor Discharge for RC=1 Capacitor")
```
